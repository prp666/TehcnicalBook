### How to use the watchdog in RaspberryPi

In raspberry pi and any other Linux system, we may facing the system is down or some service is done, if the raspberry pi is a server, it will be hard for us to reboot it face to face

To overcome this question, we need to use the tool named `watchdog`

### first we need to install the watchdog

```python
! sudo apt install watchdog
```

### According to chatgpt, we need to initialize the watchdog, bucause in RSPI the watchdog is supporting by hardware, so we can skip this step

Then the watchdog should be running automatically, but we **need to initialize the settings**

```python
! sudo nano /etc/watchdog.conf
```

This will make us can modify the config file of the watchdog

There are several important parameters, but no need to worried, the file has alot of the describtion about it

This is the mandatory setting of the watchdog, the `watchdog-device` must be specified, this is for the hardware-device.

The watchdog-timeout means if the kernel do not feed the watchdog process, after this seconds the system will reboot.

This is the value of frequently how many seconds we feed the watchdog process, suggest be smaller than `watchdog-timeout`/ 2, after feeding the dog, wait until next loop, test all the test in config, and feed the dog. 

Here is the default setting, let it been here, do not change, realtime reduce the possivility the program stuck, priority lower, more important this program is.

The `retry-timeout` is for specify, if an error continues for that time, it is an badly error, will make the process stop feeding the dog, if no this parameter, anyerror will be badly error.

This is specify the ip address, like HeartBeat protocol, each time we will sent `ping-count` time ping request to this ip address, if we receive any reply, the program will feed the dog, but if not, the system will reboot.

the `interface` is specify which interface should be active from the restart, but usually the interface is not a badly error, won't because the interface no data active cause reboot.

The pidfile is for checking if the process is alive, while linux running, we will have this kind of the file end by `pid`, if the pidfile is missing, this kind of service is not running, like the `sshd.pid` is for ssh service

### Now we can test and start the watchdog

Use `sudo systemctl enable watchdog` to start it while system is booting

Use `sudo systemctl start watchdog` to start the process now

Use `sudo systemctl disable watchdog` to disable whiel system us booting

Use `sudo systemctl stop watchdog` to stop the process, also can use for checking the watchdog can reboot well

Use `sudo systemctl status watchdog` to see the status of the watchdog
